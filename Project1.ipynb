```python
world_electricity = data[data["Indicator Name"].str.contains("Access to electricity")]
world_electricity = world_electricity[world_electricity["Indicator Name"].str.contains("rural population")]
world_electricity.dropna(how = "all")
```

```python
world_electricity = world_electricity.set_index("Country Name")
world_electricity = world_electricity.loc[:, "1990":"2014"]
```

```python
world_electricity = world_electricity.transpose().dropna(how = "all")
```

```python
world_stocks = data[data["Indicator Name"].str.contains("Stocks traded, total value")]
world_stocks = world_stocks[world_stocks["Indicator Name"].str.contains("GDP")]
world_stocks = world_stocks.dropna(how = "all")
world_stocks = world_stocks.set_index("Country Name").loc[:, "1990":"2014"]
world_stocks = world_stocks.transpose().dropna(how = "all")
```

```python
world_trade = data[data["Indicator Name"].str.contains("Trade \(")]
world_trade = world_trade.set_index("Country Name").loc[:, "1990" : "2014"].transpose()
```

```python
world_tax = data[data["Indicator Name"].str.contains("Tax revenue \(%")]
world_tax = world_tax.set_index("Country Name").loc[:, "1990":"2014"].transpose()
```

```python
array_stocks = np.asarray(world_stocks).reshape(-1)
array_electricity = np.asarray(world_electricity).reshape(-1)
array_trade = np.asarray(world_trade).reshape(-1)
array_tax = np.asarray(world_tax).reshape(-1)
```

```python
keep = ~np.isnan(array_stocks)
array_stocks = array_stocks[keep]
array_electricity = array_electricity[keep]
array_trade = array_trade[keep]
array_tax = array_tax[keep]

keep = ~np.isnan(array_electricity)
array_stocks = array_stocks[keep]
array_electricity = array_electricity[keep]
array_trade = array_trade[keep]
array_tax = array_tax[keep]

keep = ~np.isnan(array_trade)
array_stocks = array_stocks[keep]
array_electricity = array_electricity[keep]
array_trade = array_trade[keep]
array_tax = array_tax[keep]

keep = ~np.isnan(array_tax)
array_stocks = array_stocks[keep]
array_electricity = array_electricity[keep]
array_trade = array_trade[keep]
array_tax = array_tax[keep]
```

```python
data = np.array([array_stocks, array_trade, array_tax])
data = data.T
```

```python
X = sm.add_constant(data)
model = sm.OLS(array_electricity, X)
results = model.fit()
results.params
```
